```python
import rhkpy
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import holoviews as hv
import gwyfile
import scienceplots # matplotlib styles
```

```python
# Download data files from Zenodo if not cached locally
import os
import urllib.request

ZENODO_BASE = "https://zenodo.org/records/19680005/files"
DATA_DIR = "data"

def fetch_zenodo(filename):
    os.makedirs(DATA_DIR, exist_ok=True)
    local_path = os.path.join(DATA_DIR, filename)
    if not os.path.exists(local_path):
        url = f"{ZENODO_BASE}/{filename}?download=1"
        print(f"Downloading {filename} from Zenodo...")
        urllib.request.urlretrieve(url, local_path)
    return local_path
```

```python
gwy_obj = gwyfile.load(fetch_zenodo("ZrTe5_K7_300K_2019_11_04_18_02_31_072.gwy"))
channels072 = gwyfile.util.get_datafields(gwy_obj)
gwy_obj = gwyfile.load("ZrTe5K7_9K_2019_10_10_13_33_58_905.fft.gwy")
channels905 = gwyfile.util.get_datafields(gwy_obj)
```

```python
def gwy_to_hv_image(channel, label, clim=(0, 2), interp='nearest', cmap='inferno'):
    data = channel.data * 1e9
    xoff = channel.xoff * 1e9
    yoff = channel.yoff * 1e9
    xreal = channel.xreal * 1e9
    yreal = channel.yreal * 1e9
    bounds = (xoff, yoff, xoff + xreal, yoff + yreal)
    return hv.Image(data, bounds=bounds, label=label).opts(
        cmap=cmap, clim=clim,
        xaxis=None, yaxis=None,
        colorbar=True, clabel='(nm)',
        fig_size=80,
        title="",
        colorbar_opts={'extendfrac': 0},
        fontsize={'title': 12, 'labels': 10, 'xticks': 8, 'yticks': 8, 'clabel': 8, 'cticks': 8},
        interpolation=interp,
    )
```

```python
img072_topo = gwy_to_hv_image(channels072['Topography [Left]'], "072 topo", clim=(0, 1.3), interp='none', cmap='inferno')
img905_topo = gwy_to_hv_image(channels905['Topography [Left]'], "905 topo", clim=(0, 0.75), interp='none', cmap='inferno')
```

```python
img072_fft = gwy_to_hv_image(channels072['Detail 12'], '072', clim=(0, 0.5), cmap='viridis', interp='none')
img905_fft = gwy_to_hv_image(channels905['Detail 14'], '905', clim=(0, 0.5), cmap='viridis', interp='none')
```
